```python
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.by import By
import pandas as pd
```

# Set path for webdriver

```python
path='c://chromedriver.exe'
```

```python
#open browser
browser=webdriver.Chrome()
```

```python
browser.maximize_window()
```

```python
#load the webpage
browser.get('https://www.amazon.in')
```

```python
#Get rhe input element
input_search=browser.find_element(By.XPATH,"//input[@id='twotabsearchtextbox']")
button_search=browser.find_element(By.XPATH,"//input[@type='submit']")
```

```python
input_search.send_keys('Samsung phones under 10000')
sleep(1)
button_search.click()
```

# Scraping of Data

```python
products=[]
prices=[]
for i in range(10):
    print('Scraping page',i+1)
    product=browser.find_elements(By.XPATH,"//span[@class='a-size-medium a-color-base a-text-normal']")
    for p in product:
        products.append(p.text)
    price=browser.find_elements(By.XPATH,"//span[@class='a-price-whole']")
    for r in price:
        prices.append(r.text)
    next_button=browser.find_element(By.XPATH,"//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
    next_button.click()
    sleep(3)
    
```
